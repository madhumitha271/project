```python
import matplotlib.pyplot as plt

# Sample data
categories = ['department', 'age', 'region', 'gender']
values = [400, 350, 300, 450]

# Create bar chart
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='skyblue')
plt.title('employee promotion')
plt.xlabel('work time')
plt.ylabel('employee details')

# Optional: Annotate bar values
for i, val in enumerate(values):
    plt.text(i, val + 10, str(val), ha='center')

plt.tight_layout()
plt.show()
```
